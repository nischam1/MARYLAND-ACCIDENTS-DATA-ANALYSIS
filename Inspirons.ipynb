##Person Details Dataset

##Vehicle Details Dataset

```python
#@title Reading Vehicle Details Dataset
vehicle_details = pd.read_csv("https://media.githubusercontent.com/media/sravansaii/marylanddata/main/Maryland_Statewide_Vehicle_Crashes_-_Vehicle_Details.csv")# Reading Vehicle details data set
print(vehicle_details)
```

```python
#@title Removing unnecessary columns from Vehicle Details Dataset
vehicle_details.drop(['CONTI_DIRECTION_DESC',
       'CONTI_DIRECTION_CODE','VIN_NO','CV_BODY_TYPE_DESC',
       'CV_BODY_TYPE_CODE','VEH_MODEL',
       'COMMERCIAL_FLAG_DESC', 'COMMERCIAL_FLAG', 'HZM_NUM',
       'TOWED_AWAY_FLAG_DESC', 'TOWED_AWAY_FLAG', 'NUM_AXLES', 'GVW_DESC',
       'GVW_CODE', 'GOING_DIRECTION_DESC', 'GOING_DIRECTION_CODE','DRIVERLESS_FLAG_DESC',
       'DRIVERLESS_FLAG', 'FIRE_FLAG_DESC', 'FIRE_FLAG', 'PARKED_FLAG_DESC',
       'PARKED_FLAG','HAZMAT_SPILL_FLAG_DESC', 'HAZMAT_SPILL_FLAG','TOWED_VEHICLE_CONFIG_DESC', 'TOWED_VEHICLE_CONFIG_CODE',
       'AREA_DAMAGED_CODE_IMP1', 'AREA_DAMAGED_CODE1', 'AREA_DAMAGED_CODE2',
       'AREA_DAMAGED_CODE3',],inplace=True,axis=1)   #dropping all the irrelevant columns which is not necesary in the data analysis
```

##Filtering And Merging of Datasets

```python
#@title Filtering 2020 and 2021 Data from Vehicle Crashes Dataset
crashes_2020_2021 = vehicle_crashes[vehicle_crashes.YEAR >= 2020] #Getting only 2020 and 2021 data
```

```python
#@title Filtering 2020 and 2021 Data from Person Details Dataset
people_2020_2021 = personality_data[personality_data.YEAR >= 2020]  #Getting only 2020 and 2021 data
```

```python
#@title Filtering 2020 and 2021 Data from Vehicle Details Dataset
Vdetails_2020_2021 = vehicle_details[vehicle_details.YEAR >= 2020] #Getting only 2020 and 2021 data
```

```python
#@title Merging Vehicle Crashes and Person Details Datasets
dataset_merge1 = pd.merge(crashes_2020_2021, people_2020_2021, on ='REPORT_NO',how = 'left') #merging 2 datasets
dataset_merge1
```

```python
#@title Merging Vehicles Details Dataset with previously merged dataset
filtered_dataset = pd.merge(dataset_merge1, Vdetails_2020_2021, on ='REPORT_NO',how = 'left') #Merging Vehicles Details Dataset with previously merged dataset
filtered_dataset
```

```python
#@title Converting Accident Date format and Assigning New Column in Merged Dataset
filtered_dataset['accident_date'] =  pd.to_datetime(filtered_dataset['ACC_DATE'], format='%Y%m%d') #creating new column accident_date and loading the proper date format in that column
filtered_dataset
```

```python
#@title Droping Unwanted Columns in Merged Data
Cleaned_dataset = filtered_dataset.drop(['YEAR_y', 'YEAR', 'PERSON_TYPE', 'Quarter_x', 'Quarter_y'], axis = 1) #dropping repeated columns
Cleaned_dataset
```

##Data Visualization

###Visualization for combined data of 2020 and 2021
